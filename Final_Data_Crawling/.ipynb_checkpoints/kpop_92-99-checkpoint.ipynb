# 1. 92~99년도 데이터 합치기

```python
pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
```

```python
df_1= pd.read_csv('kpop_92-93.tsv', encoding = 'utf-8',sep='\t',header=None,names=["년도","가수","제목","성별","장르","최고순위","작사가","작곡가","소속사","가사"])
df_2= pd.read_csv('kpop_94-95.tsv', encoding = 'utf-8',sep='\t',header=None,names=["년도","가수","제목","성별","장르","최고순위","작사가","작곡가","소속사","가사"])
df_3= pd.read_csv('kpop_96-97.tsv', encoding = 'utf-8',sep='\t',header=None,names=["년도","가수","제목","성별","장르","최고순위","작사가","작곡가","소속사","가사"])
df_4= pd.read_csv('kpop_98-99.tsv', encoding = 'utf-8',sep='\t',header=None,names=["년도","가수","제목","성별","장르","최고순위","작사가","작곡가","소속사","가사"])
```

```python
df_1["년도"]=df_1["년도"].str.replace(".","")
df_4["년도"]=df_1["년도"].str.replace(".","")
```

```python
df_1.astype({"년도":'int64'})
df_4.astype({"년도":'int64'})
```

```python
kpop_all=pd.concat([df_1,df_2,df_3,df_4])
```

```python
kpop_all.dtypes
```

```python
kpop_all.head()
```

```python
kpop_all=kpop_all.astype({'년도':str}) #년도를 문자열로 변경
```

```python
def append_zero(x): #yyyymmdd 형식으로 변경: 월,일이 누락된 경우 yyyy0000, yyyymm00 로 만들어줌
    if len(x)==4:
        x=x+'0000'
        return x
    elif len(x)==6:
        x=x+'00'
        return x
    else:
        return x
```

```python
kpop_all["년도"]=kpop_all["년도"].apply(append_zero)
```

```python
kpop_all.head()
```

```python
kpop_all=kpop_all.astype({'년도':'int64'})    #type int로 변경
kpop_all=kpop_all.astype({'최고순위':'int64'})
```

```python
kpop_all=kpop_all.fillna('NULL')  #결측치 -> 순위:0 ,나머지:NULL
```

```python
kpop_all.dtypes
```

# 2. 합친 92-99년도 tsv 파일로 저장

```python
kpop_all.to_csv('kpop_92-99.tsv',index=False,header=None,sep='\t')
```
